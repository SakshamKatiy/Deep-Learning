```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (6,6))
```

```python
def Sigmoid_b(x):
    return (1/(1+np.exp(-x)))
x = np.linspace(-5,5)
plt.plot(x,Sigmoid_b(x))
plt.show()
```

```python
def TanH(x):
    return np.tanh(x)
plt.title('Activation TanH')
plt.plot(x,TanH(x))
plt.show()
```

```python
def RectifiedLinearUnit(x):
    list = []
    for i in x:
        if i>=0:
            list.append(i)
        else:
            list.append(0)
    return list
x = np.linspace(-115,500)
plt.plot(x,RectifiedLinearUnit(x))
plt.show
```

```python
def leakyrelu(x):
    list = []
    for i in x:
        if i>=0:
            list.append(i)
        else:
            list.append(0.01*i)
    return list
plt.plot(x,leakyrelu(x))
plt.show()
```

```python
def softmax(x):
    return (np.exp(x)/np.sum(np.exp(x)))
plt.plot(x,softmax(x))
plt.title('Softmax function')
plt.show()
```
